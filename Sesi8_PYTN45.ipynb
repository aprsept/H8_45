```python
canada = pd.read_excel("./data/Canada.xlsx", sheet_name="Canada by Citizenship",
                         skiprows=range(20), skipfooter=2)
print("num of rows:", canada.shape[0])
print("num of cols:", canada.shape[1])

canada
```

```python
canada.shape
```

```python
canada.drop(["AREA", "REG", "DEV", "Type", "Coverage"], axis=1, inplace=True)

canada.rename(columns={
    "OdName": "Country",
    "RegName": "Region",
    "AreaName": "Continent"
    },
    inplace=True
)

canada.set_index("Country", inplace=True)

canada["Total"] = canada.sum(axis=1)
years = list(range(19980, 2014))

canada.head()
```

```python
max_words = 90

word_string = ""
for country in canada.index:
    num_repeat = int(canada.loc[country, "Total"] / float(canada.Total.sum()) * max_words)
    word_string = word_string + ((country + " ") * num_repeat)
    
word_string
```

```python
country_wordcloud = WordCloud(background_color="lavenderblush").generate(word_string)

plt.figure(figsize=(11, 6))
plt.imshow(country_wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show
```

# Regression Plots using Seaborn

```python
years = list(range(1980, 2014))
df_total = pd.DataFrame(canada[years].sum(axis=0))
df_total.index = map(float, df_total.index)
df_total.reset_index(inplace=True)
df_total.columns = ["years", "total"]
df_total.head()
```

```python
sns.regplot(df_total["years"], df_total["total"])
```

```python
plt.figure(figsize=(15, 10))

sns.set(font_scale=1.5)
ax = sns.regplot(x="years", y="total", data=df_total, color="darkorchid", marker="+", scatter_kws={"s": 200})

plt.title("Immigration to Canada (1980-2014)")
plt.xlabel("Years")
plt.ylabel("Total Immigrants")
plt.show()
```

```python
incidents = pd.read_csv("data/incidents.csv")
print("shape:", incidents.shape)
incidents.head()
```

```python
sample_incidents = incidents.sample(n=100, random_state=1)
sample_incidents.head()
```

```python
# SF latitude and longitude
latitude = 37.77
longitude = -122.42

sf_map = folium.Map(location=(latitude, longitude), zoom_start=12)
sf_map
```

```python
incidents_loc = folium.map.FeatureGroup()

for lat, lng in zip(sample_incidents.Y, sample_incidents.X):
    incidents_loc.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            color="tomato",
            fill=True,
            fill_color="lightcyan",
            fill_opacity=.9
        )
    )
    
latitude = list(sample_incidents.Y)
longitude = list(sample_incidents.X)
category = list(sample_incidents.Category)
for lat, lng, cat in zip(latitude, longitude, category):
    folium.Marker([lat, lng], popup=cat).add_to(sf_map)
    
    
sf_map.add_child(incidents_loc)
```

```python
import json
with open('data/world-countries.json') as data_file:
    data = json.load(data_file)

canada.reset_index(inplace=True)

world_geo = data
world_map = folium.Map(location=[0, 0], zoom_start=2, tiles = 'Mapbox Bright')
world_map.choropleth(
    geo_data = world_geo,
    data = canada,
    columns = ['Country', 'Total'],
    key_on = 'feature.properties.name',
    fill_color = 'YlOrRd', 
    fill_opacity = 0.7, 
    line_opacity = 0.2,
    legend_name = 'Immigration to Canada'
)

world_map
```
