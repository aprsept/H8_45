# Inferential Statistics

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm

# for latex equations
from IPython.display import Math, Latex

# for displaying images
from IPython.core.display import Image

from scipy.stats import uniform, norm, gamma, expon, poisson, binom, bernoulli

#seaborn configuration
sns.set(color_codes=True)
sns.set(rc={"figure.figsize": (5, 5)})
```

```python
def plot_distribution(data, color, title):
    ax = sns.displot(
        data,
        bins=100,
        kde=True,
        color=color,
        linewidth=15,
        alpha=1
    )
    ax.set(xlabel=title, ylabel="Frequency")
    plt.title(title)
    plt.show()
```

# Uniform Distribution

```python
n = 100
loc = 10
scale = 20
rv_uniform = uniform.rvs(loc=loc, scale=scale, size=n)

# plot distribution
sns.distplot(rv_uniform,
            bins=10,
            color="MediumPurple",
            hist_kws={"linewidth": 15, "alpha": 1})
```

```python
# plot distribution
ax = sns.displot(rv_uniform,
                color="MediumSeaGreen",
                linewidth=15, 
                alpha=1,
                kde=True,
                bins=100)

ax.set(xlabel="Uniform Distribution", ylabel="Frequency")
plt.title("Uniform Distribution")
plt.show()
```

# Confidence Intervals for Proportion

```python
heart = pd.read_csv("data/heart.csv")
print("Dataset shape:", heart.shape)
display(heart.head())
```

```python
cross_tab = pd.crosstab(heart.target, heart.sex)
cross_tab["Total_target"] = cross_tab["Female"] + cross_tab["Male"]
cross_tab
```

```python
pop_female = 312
pop_male = 713

prop_female = 226 / pop_female
print("Proportion of female with disease:", prop_female)
```

```python
stderr_female = np.sqrt(prop_female * (1-prop_female) / pop_female)
print("Standard error of female proportion with disease:", stderr_female)
```

```python
def compute_CI(estimate, stderr, z_score=1.96):
    return (
        estimate - z_score*stderr,
        estimate + z_score*stderr
    )
```

```python
lower_ci, upper_ci = compute_CI(prop_female, stderr_female)
print(f"CI for female proportion with disease: [{lower_ci:.3f}, {upper_ci:.3f}]",)
```

```python
sm.stats.proportion_confint(pop_female * prop_female, pop_female)
```
