```python
arr_int = np.array([1, 2, 3, 4, 5, 6])
print(arr_int, arr_int.shape)
```

```python
series_int = pd.Series(data=arr_int, index=['a', 'b', 'c', 'd', 'e', 'f'], dtype="float16", name="floats")
print(series_int)
print(series_int.index)

series_floats = pd.Series(data=arr_int,dtype="float16", name="floats")
print(series_floats)

series_members = pd.Series(
    data=["namjoon", "taehyung", "hoseok", "yoongi", "jimin", "jungkook", "seokjin"],
    name="members"
)
print(series_members)
print(series_members.index)
```

```python
pd.DataFrame(['a', 'b', 'c', 'd', 'e', 'f'])
```

```python
pd.DataFrame(['a', 'b', 'c', 'd', 'e', 'f']).shape
```

```python
pd.DataFrame([[1, 2, 3], [4, 5, 6]], columns=["a", "b", "c"], index=["aa", "bb"])
```

```python
data = pd.DataFrame({"floats": series_floats, "members": series_members})
```

```python
pd.DataFrame([series_floats, series_members]).T
```
