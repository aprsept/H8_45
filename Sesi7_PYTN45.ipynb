```python
indonesia = df_canada.loc[["Indonesia"], years].T
indonesia
```

```python
indonesia.plot()
plt.title("Immigrants from Indonesia")
plt.xlabel("Years")
plt.ylabel("Number of Immigrants")
plt.text(1993, 1000, "1998 Crisis", fontsize=10)
plt.show()
```

```python
imci = df_canada.loc[["Indonesia", "Malaysia", "China", "India"], years]
imci
```

```python
imci.T.plot(figsize=(15, 6))
plt.title("Immigrants from Indonesia, Malaysia, China, India")
plt.xlabel("Years")
plt.ylabel("Number of Immigrants")
```

Top 5 countries that contribute the most to Canada immigrants

```python
top_5_countries = df_canada.sort_values(by="Total", ascending=False).iloc[:5][years].T
top_5_countries
```

```python
top_5_countries.plot(figsize=(12, 7))
plt.title("Top 5 Countries")
plt.xlabel("Years")
plt.ylabel("Number of Immigrants")
plt.show()
```

```python
top_5_countries.plot(kind="area",figsize=(12, 7), stacked=False, alpha=.35, 
                     color=["teal", "sandybrown", "olivedrab", "indigo", "tomato"])
plt.title("Top 5 Countries")
plt.xlabel("Years")
plt.ylabel("Number of Immigrants")
plt.savefig("./data/top5countries.png")
plt.show()

```

```python
df_canada[2013].plot(kind="hist", figsize=(12, 6), color="darkcyan")
plt.title("Immigrants Distribution in 2013")
plt.xlabel("#Immigrants")
plt.ylabel("#Countries")
plt.show()
```

```python
dns = df_canada.loc[["Denmark", "Norway", "Sweden"], years].T
dns
```

```python
dns.plot(
    kind="hist",
    figsize=(14, 7),
    alpha=.6,
    bins=15,
    color=["darkviolet","lime","salmon"]
)
plt.title("Immigrants Distribution from Denmark, Norway, Sweden")
plt.xlabel("#Immigrants")
plt.ylabel("#Countries")
plt.show()
```

```python
iceland = df_canada.loc["Iceland", years].T
iceland
```

```python
iceland.plot(
    kind="bar",
    figsize=(14, 7),
    color="gold"
)
plt.title("Immigrants Distribution from Iceland")
plt.xlabel("Year")
plt.ylabel("#Immigrants")
plt.show()
```

```python
iceland.plot(
    kind="barh",
    figsize=(14, 7),
    color="lightskyblue"
)
plt.title("Immigrants from Iceland (1980-2013)")
plt.xlabel("Year")
plt.ylabel("#Immigrants")
plt.show()
```

```python
iceland.plot(
    kind="bar",
    figsize=(14, 7),
    color="lightsalmon"
)
plt.title("Immigrants from Iceland (1980-2013)")
plt.xlabel("Year")
plt.ylabel("#Immigrants")
plt.annotate("", xy=(30, 70), xytext=(28, 20), xycoords="data",
            arrowprops={"arrowstyle": "->",
                       "connectionstyle": "arc3",
                       "color": "limegreen", "lw": 2})
plt.annotate("2008-2011 Financial Crises", xy=(28, 30), rotation=80, va="bottom", ha="left")
plt.show()
```

```python
continents = df_canada.groupby("Continent").sum()

pie_colors = ["chartreuse", "indianred", "powderblue", "goldenrod", "mediumorchid", "tomato"]
explode_list = [.1, 0, 0, 0, .1, .1]
continents["Total"].plot(kind="pie", figsize=(20, 6), shadow=True, autopct="%1.1f%%", 
                         explode=explode_list,
                        colors=pie_colors, pctdistance=1.15, labels=None)
plt.legend(labels=continents.index, loc="lower right")
```

```python
df_canada.sort_values(by="Total", ascending=True, inplace=True)
top_10 = df_canada["Total"].tail(10)
top_10
```

```python
top_10.plot(kind="barh", figsize=(10, 11), 
            color=["steelblue","mediumorchid", "yellow", "hotpink", "GreenYellow", "MediumAquamarine", "DarkGoldenrod",
                  "LavenderBlush", "CadetBlue", "LightCoral"])
plt.xlabel("Number of Immigrants")
plt.ylabel("Country")
plt.title("Top 10 Countries Contributed of The Most Immigrants in Canada 1980-2013")
plt.show()
```

```python
id_ma = imci.loc[["Indonesia", "Malaysia"]]
cn_in = imci.loc[["China", "India"]]
id_ma
```

```python
fig, axes = plt.subplots(2, 1, figsize=(12, 6))
for df, ax in zip((id_ma, cn_in), axes):
    ax.set_title(
        f"{' and '.join(df.index.tolist())} Immigrants"
    )
    df.T.plot(kind="line", ax=ax)
```
